```python
df3 = pd.read_csv('C://Users//harsh//Desktop//isro hacakthon//data//raw//Delhi_wardwise_lst_and_ndvi_and_builtup_pop_2018_to_2025.csv')
```

```python
df3.head()
```

```python
df3 = df3.drop(columns = ['Unnamed: 0.2','Unnamed: 0.1','Unnamed: 0','method','method_builtpct','pixel_count_ndvi','valid_pixel_count_ndvi','cloud_pixel_percent','pixel_count_lst','valid_pixel_count_lst','invalid_pixel_percent'])
```

```python
df3.head()
```

```python
df3 = df3.drop(columns = ['method_ndvi_and_lst'])
```

```python
df3.head()
```

```python
df3['mean_lst_day_celsius'].describe()
```

```python
q1 = 22.974194
q3 = 41.312901
median_val = 33.353567

iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
df3["mean_lst_day_celsius"] = np.where(
    (df3["mean_lst_day_celsius"] < lower_bound) | (df3["mean_lst_day_celsius"] > upper_bound),
    median_val,
    df3["mean_lst_day_celsius"],
)
```

```python
df3['mean_lst_day_celsius'].describe()
```

```python
df3['month_sin'] = np.sin((2*3.14*df3['month'])/12)
```

```python
df3['month_cos'] =  np.cos((3*3.14*df3['month'])/12)
```

```python
df3.head()
```

```python
df3 = df3.sort_values(['zone','ward','year','month'])
```

```python
df3.head()
```

```python
df3['lst_lag1'] = df3.groupby('ward')['mean_lst_day_celsius'].shift(1)
```

```python
df3['lst_lag2'] = df3.groupby('ward')['mean_lst_day_celsius'].shift(2)
```

```python
df3['lst_lag3'] = df3.groupby('ward')['mean_lst_day_celsius'].shift(3)
```

```python
df3.head()
```

```python
df3['lst_3month_avg'] = (df3.groupby("ward")["mean_lst_day_celsius"].shift(1).rolling(3).mean())
```

```python
df3['lst_12month_avg'] = (df3.groupby("ward")["mean_lst_day_celsius"].shift(1).rolling(12).mean())
```

```python
df3["ndvi_lag1"] = (df3.groupby("ward")["mean_ndvi"].shift(1))
```

```python
df3["ndvi_lag2"] = (df3.groupby("ward")["mean_ndvi"].shift(2))
```

```python
df3["ndvi_3month_avg"] = (df3.groupby("ward")["mean_ndvi"].shift(1).rolling(3).mean())
```

```python
df3["ndvi_change"] = (df3.groupby("ward")["mean_ndvi"].diff())
```

```python
df3["builtup_ndvi"] = (df3["built_up_percent"]*df3["mean_ndvi"])
```

```python
df3["urban_intensity"] = (df3["population_density"]*df3["built_up_percent"])
```

```python
df3["heat_exposure_index"] = (df3['population_density']*df3['built_up_percent'])
```

```python
df3["year_index"] = (df3['year'] - 2018)
```

```python
df3 = df3.drop(columns = ['mean_lst_temp'])
```
