```python
df1 = pd.read_csv('../data/raw/sorted_lst_for_MP.csv')
df2 = pd.read_csv('../data/raw/sorted_NDVI_for_MP.csv')
df_merged_1 = pd.merge(df1, df2, on=['district_code','year','month','district'], how='right')
df_merged_1[['year', 'month']] = df_merged_1[['year','month']].astype(int)
```

```python
monthly_day_avg = (df_merged_1.groupby(["district_code","month"])['lst'].transform("mean"))
df_merged_1['lst'] = df_merged_1['lst'].fillna(monthly_day_avg)
```

```python
monthly_night_avg = (df_merged_1.groupby(["district_code","month"])['night_lst'].transform("mean"))
df_merged_1['night_lst'] = df_merged_1['night_lst'].fillna(monthly_night_avg)
```

```python
import numpy as np
df3 = pd.read_csv('../data/raw/sorted_pop_for_MP.csv')
districts = df3[['district_code','district']].drop_duplicates()
years = list(range(2015,2026))
grid = []
for _, row in districts.iterrows():
    for year in years:
        grid.append({
            'district_code': row['district_code'],
            'district': row['district'],
            'year': year
        })
full_grid_df = pd.DataFrame(grid)
df3_expanded = pd.merge(full_grid_df,df3,on=['district_code','district','year'], how='left')
df3_expanded['total_population'] = (df3_expanded.groupby('district_code')['total_population'].transform(lambda x: x.interpolate(method='linear')))
```

```python
df4 = pd.read_csv('../data/raw/sorted_builtup_pct_MP.csv')
districts = df4[['district_code','district']].drop_duplicates()
years = list(range(2015,2026))
grid = []
for _, row in districts.iterrows():
    for year in years:
        grid.append({
            'district_code': row['district_code'],
            'district': row['district'],
            'year': year
        })
full_grid_df1 = pd.DataFrame(grid)
df4_expanded = pd.merge(full_grid_df1,df4, on = ['district_code','district','year'], how = 'left')
df4_expanded['built_up_area_sq_m'] = (df4_expanded.groupby('district_code')['built_up_area_sq_m'].transform(lambda x: x.interpolate(method='linear')))
df4_expanded.head(20)
```

```python
df11 = pd.read_csv('../data/final/final_mp_dataset.csv')
```

```python
df11.drop(columns = ['Unnamed: 0.1','Unnamed: 0'], inplace=True)
```

```python
df11.head()
```

```python
df11['ndvi_builtup_interaction'] = df11['ndvi']*df11['built_up_pct']
```

```python
df11['population_density'] = df11['total_population']/df11['district_area_sq_km']
```

```python
df11['built_up_growth'] = df11.groupby('district')['built_up_pct'].diff()
```

```python
df11['built_up_growth'] = df11['built_up_growth'].fillna(0)
```

```python
df11['ndvi_annual_average'] = df11.groupby(['district','year'])['ndvi'].transform('mean')
```

```python
df11['ndvi_anomally'] = df11['ndvi'] - df11['ndvi_annual_average']
```

```python
df11['district_avg_lst'] = df11.groupby('district')['lst'].transform('mean')
```

```python
df11['lst_anomally'] = df11['district_avg_lst']-df11.groupby('district')['lst'].shift(1)
```
